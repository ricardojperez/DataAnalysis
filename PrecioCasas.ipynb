```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(df_train["GrLivArea"].values.reshape(-1,1),df_train["SalePrice"])
```

```python
lin_reg.intercept_
```

```python
lin_reg.coef_
```

```python
X= df_train["GrLivArea"]
y = df_train["SalePrice"]
```

```python
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1234)
```

```python
plt.scatter(X_test, y_test)
plt.show()
```

```python
Xmin_max = np.array([[df_train["GrLivArea"].min()],[df_train["GrLivArea"].max()]])
```

```python
y_train_pred= lin_reg.predict(Xmin_max)
```

```python
plt.plot(Xmin_max,y_train_pred,"g-")
plt.plot(df_train["GrLivArea"],df_train["SalePrice"],"b.")
plt.xlabel("Area de vivienda")
plt.ylabel("Costo de vivienda")
plt.show()
```

```python
Y_pred=lin_reg.predict(X_test.values.reshape(-1,1))
```

```python
plt.plot(Xmin_max,y_train_pred,"g-")
plt.plot(X_test,Y_pred,"b.")
plt.xlabel("Area de vivienda")
plt.ylabel("Costo de vivienda")
plt.show()
```
